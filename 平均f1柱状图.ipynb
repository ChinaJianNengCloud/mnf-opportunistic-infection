```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 读取Excel文件
file_path = '/Users/louhao/Desktop/十六支队新增或有补充的图片1008/10-5测试集平均F1柱状图/shap与permutation10-5测试集平均F1.xlsx'  # 替换为你的Excel文件路径
sheet_name = 'permutation'  
df = pd.read_excel(file_path, sheet_name=sheet_name)

# 根据第二列（数值列）进行降序排序
df_sorted = df.sort_values(by=df.columns[1], ascending=False)

# 提取排序后的Model和数值
models = df_sorted[df.columns[0]]
values = df_sorted[df.columns[1]]
```

```python
# 绘制柱状图
plt.figure(figsize=(10, 6))
bars = plt.bar(models, values, color='skyblue')
for bar, value in zip(bars, values):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{value:.4f}', 
             ha='center', va='bottom', fontsize=10)  # 显示数值
#plt.title('Model vs Value', fontsize=14)
#plt.xlabel('Model', fontsize=12)
plt.ylabel(f'{sheet_name}_mean_f1', fontsize=12)
plt.xticks(rotation=45, ha='right')  # 旋转x轴标签以避免重叠
plt.tight_layout()  # 自动调整子图参数以避免重叠

# 保存图片
plt.savefig(f'/Users/louhao/Desktop/十六支队新增或有补充的图片1008/10-5测试集平均F1柱状图/{sheet_name}.png', dpi=300)  # 将图片保存为PNG文件，分辨率为300dpi

# 显示图形
plt.show()
```
