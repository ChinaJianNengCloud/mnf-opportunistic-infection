```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Excel
file_path = '/Users/louhao/Desktop/10-5/10-5.xlsx'
sheet_name = 'shap+permutation'
df = pd.read_excel(file_path, sheet_name=sheet_name)
# 
df = df.sort_values(by=df.columns[1], ascending=False)
x = df.columns[1:]
y_labels = df.iloc[:, 0]

plt.figure(figsize=(14, 12))

for i, y_label in enumerate(y_labels):
    y = df.iloc[i, 1:]
    plt.plot(x, y, label=y_label)

# 
#plt.xlabel('shap10-5')  
plt.ylabel('frequency')  
plt.title('shap+permutation10-5')  
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))  # 
plt.grid()
plt.savefig('/Users/louhao/Desktop/10-5/shap+permutation10-5.png', dpi=300)
# 
plt.show()
```
