```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# 读取Excel文件
file_path = '/Users/louhao/Desktop/（新增图片）不同特征在不同模型前10-5出现频次/不同特征在不同模型中前10-5排名出现的频数.xlsx'
sheet_name = 'shap+permutation'
df = pd.read_excel(file_path, sheet_name=sheet_name)
# 以第一行作为横坐标，第一列作为纵坐标
df = df.sort_values(by=df.columns[1], ascending=False)
x = df.columns[1:]
y_labels = df.iloc[:, 0]

plt.figure(figsize=(14, 12))

for i, y_label in enumerate(y_labels):
    y = df.iloc[i, 1:]
    plt.plot(x, y, label=y_label)

# 添加图例、标题和轴标签
#plt.xlabel('shap10-5')  
plt.ylabel('frequency')  
plt.title('shap+permutation10-5')  
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))  # 将图例放到右边
plt.grid()
plt.savefig('/Users/louhao/Desktop/（新增图片）不同特征在不同模型前10-5出现频次/shap+permutation10-5.png', dpi=300)
# 显示图形
plt.show()
```
