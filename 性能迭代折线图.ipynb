```python
import pandas as pd
import matplotlib.pyplot as plt

# Excel
file_path = '/Users/louhao/Desktop/161008//3210-56AUCACCF1/.xlsx' 
sheet_name = 'permutation'  
df = pd.read_excel(file_path, sheet_name=sheet_name)

# 
x_labels = df.iloc[:, 0]  # x
auc_values = df.iloc[:, 1]  # AUC
acc_values = df.iloc[:, 2]  # ACC
f1_values = df.iloc[:, 3]  # F1

# 
plt.figure(figsize=(10, 6))

# 
plt.plot(x_labels, auc_values, label='AUC', marker='o', color='b')
plt.plot(x_labels, acc_values, label='ACC', marker='o', color='g')
plt.plot(x_labels, f1_values, label='F1', marker='o', color='r')

# 
plt.title(sheet_name, fontsize=14)
#plt.xlabel('X Labels', fontsize=12)
plt.ylabel('Metric Values', fontsize=12)

# 
plt.legend(loc='best')

# x
plt.xticks(rotation=45, ha='right')

# 
plt.tight_layout()

# 
plt.savefig(f'/Users/louhao/Desktop/161008//3210-56AUCACCF1/{sheet_name}.png', dpi=300)

# 
plt.show()
```
