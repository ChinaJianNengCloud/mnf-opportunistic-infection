```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取Excel文件
file_path = '/Users/louhao/Desktop/16支队新增或有补充的图片1008/拼图/32特征、10-5特征的最优模型以及双方组合的6特征的性能（AUC、ACC、F1）/数据.xlsx' 
sheet_name = 'permutation'  
df = pd.read_excel(file_path, sheet_name=sheet_name)

# 提取数据
x_labels = df.iloc[:, 0]  # 第一列作为x轴标签
auc_values = df.iloc[:, 1]  # 第二列作为AUC数据
acc_values = df.iloc[:, 2]  # 第三列作为ACC数据
f1_values = df.iloc[:, 3]  # 第四列作为F1数据

# 创建图形
plt.figure(figsize=(10, 6))

# 绘制三条折线
plt.plot(x_labels, auc_values, label='AUC', marker='o', color='b')
plt.plot(x_labels, acc_values, label='ACC', marker='o', color='g')
plt.plot(x_labels, f1_values, label='F1', marker='o', color='r')

# 添加标题和轴标签
plt.title(sheet_name, fontsize=14)
#plt.xlabel('X Labels', fontsize=12)
plt.ylabel('Metric Values', fontsize=12)

# 添加图例
plt.legend(loc='best')

# 显示x轴标签的角度，防止重叠
plt.xticks(rotation=45, ha='right')

# 自动调整子图布局，避免重叠
plt.tight_layout()

# 保存图片
plt.savefig(f'/Users/louhao/Desktop/16支队新增或有补充的图片1008/拼图/32特征、10-5特征的最优模型以及双方组合的6特征的性能（AUC、ACC、F1）/{sheet_name}.png', dpi=300)

# 显示图形
plt.show()
```
